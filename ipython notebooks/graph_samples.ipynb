```python
import matplotlib.pyplot as plt
import networkx as nx


G=nx.Graph()

"""
Python SQL
Python PostgreSQL MongoDB C/C++
Python PostgreSQL C/C++
Python SQL Linux Git Bash
Python PostgreSQL Bash Django Git
Python SQL PostgreSQL MongoDB
Python MongoDB
"""

weighted_pairs = [("python", "SQL", 2),
                  ("python", "PostgreSQL", 4),
                  ("python", "python", 4),
                  ("python", "MongoDB", 3),
                  ("python", "C/C++", 2),
                  ("python", "Linux", 4),
                  ("python", "Git", 2),
                  ("python", "Bash", 2),
                  ("python", "Django", 1),
                  ("PostgreSQL", "MongoDB", 2),
                  ("PostgreSQL", "C/C++", 2),
                  ("PostgreSQL", "Bash", 1),
                  ("PostgreSQL", "Django", 1),
                  ("PostgreSQL", "Git", 1),
                  ("PostgreSQL", "SQL", 1),
                  ("MongoDB", "C/C++", 1),
                  ("MongoDB", "SQL", 1),
                  ("MongoDB", "C/C++", 1),
                  ("SQL", "Linux", 1),
                  ("SQL", "Git", 1),
                  ("SQL", "Bash", 1),
                  ("Git", "Linux", 1),
                  ("Git", "Bash", 1),
                  ("Git", "Django", 1),
                  ("Django", "Bash", 1),
                  ("Git", "Bash", 1),
                  ("Git", "Linux", 1),
                 ]

G.add_weighted_edges_from(weighted_pairs)
pos = nx.spring_layout(G)
weights = [e[2] for e in weighted_pairs]
print(list(G.nodes()))
all_words = [w for e in weighted_pairs for w in e[:2]]
sizes = [(all_words.count(i)**2)*100 for i in list(G.nodes())]
print([all_words.count(i) for i in list(G.nodes())])

plt.figure(3,figsize=(12,12))

nx.draw_networkx_nodes(G, pos, node_color='#A0CBE2', node_size=sizes, node_cmap=plt.cm.Blues)
nx.draw_networkx_edges(G, pos, node_color='#A0CBE2', width=weights, edge_cmap=plt.cm.Blues)

nx.draw_networkx_labels(G, pos)

plt.show()
```
